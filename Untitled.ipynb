```python
import torch

for i in range(11):
    segment_name = f"{i:05d}.pt"
    result_1 = torch.load(f"data/pgto/{segment_name}")
    result_2 = torch.load(f"data/pgto_cpu/{segment_name}")
    result_3 = torch.load(f"data/pgto_cpu_1/{segment_name}")
    
    best_result = min(result_1["best_cost"], result_2["best_cost"], result_3["best_cost"])
    
    print(f"segment ({segment_name}), best_result: {best_result}")
```

```python
from offline.gdrive import GDriveManager

gdrive = GDriveManager(
    root_folder_name="pgto"
)
```

```python
import os
from pathlib import Path
from tqdm import tqdm

data_path = "data/pgto"
for file_name in tqdm(sorted(os.listdir(data_path))):
    if file_name.endswith(".pt"):
        if not gdrive.file_exists(file_name):
            gdrive.upload_file(Path(os.path.join(data_path, file_name)))
```
