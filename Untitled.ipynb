```python
import torch

for i in range(11):
    segment_name = f"{i:05d}.pt"
    result_1 = torch.load(f"data/pgto/{segment_name}")
    result_2 = torch.load(f"data/pgto_cpu/{segment_name}")
    result_3 = torch.load(f"data/pgto_cpu_1/{segment_name}")
    
    best_result = min(result_1["best_cost"], result_2["best_cost"], result_3["best_cost"])
    
    print(f"segment ({segment_name}), best_result: {best_result}")
```
